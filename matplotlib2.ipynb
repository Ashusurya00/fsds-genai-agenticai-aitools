```python
#error bar chart
x9 = np.arange(0,4,0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr = e1, fmt = '.-')
plt.show()
```

```python
#stacked bar chart 
a = [15.,30.,45.,22.]

b = [15.,25.,50.,20.]

z2 = range(4)
plt.bar(z2,a,color='b')
plt.bar(z2,b,color='r',bottom=a)
plt.show()
```

```python
#pie chart
plt.figure(figsize=(7,7))

x10 = [35,25,20,20]
labels = ['computer','electronic','mechanical','chemical']
plt.pie(x10, labels=labels);
plt.show()
```
