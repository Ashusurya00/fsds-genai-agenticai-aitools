```python
import pandas as pd

# Data
name = ["Ram", "Shyam", "Aditi", "Jaggu"]
age = [20, 22, 21, 23]
marks = [85, 90, 78, 92]

df = pd.DataFrame({
    "name": name,
    "age": age,
    "marks": marks
})

# Function to assign grades
def assign_grade(m):
    if m >= 90:
        return "A"
    elif m >= 80:
        return "B"
    else:
        return "C"

# Add new column
df["grade"] = df["marks"].apply(assign_grade)

print(df)
```

```python
import pandas as pd

# Employee data
data = {
    "dept": ["IT", "HR", "IT", "Finance", "HR"],
    "salary": [40000, 30000, 45000, 50000, 32000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Find average salary per department
avg_salary = df.groupby("dept")["salary"].mean()

print("\nAverage salary per department:")
print(avg_salary)
```
