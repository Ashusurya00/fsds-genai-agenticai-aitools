```python
import matplotlib.pyplot as plt

x = ['mon','tue','wed','thu','fri']
y = [10,15,7,20,12]

plt.plot(x,y)

plt.title('bakery sales this week')

plt.xlabel('days of the week')
plt.ylabel('sales per day')

plt.show()
```
