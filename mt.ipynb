```python
import matplotlib.pyplot as plt
```

```python
months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months, sales, color='Blue', linestyle='--', linewidth=2, marker='o' , label='2025 sales data')
plt.xlabel('months')
plt.ylabel('sales per month')
plt.title('monthly sales data report')
plt.legend()
plt.grid(color = 'gray',linestyle=':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['m1','m2','m3','m4'])
```

```python
product = ['a','b','c','d']

sales = [1000,1500,800,1200]

plt.bar(product,sales,color='orange',label='sales 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product sale comparison')
plt.legend()
plt.show()
```

```python
regions = ['north','south','east','west']
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels = regions, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.title('reveue across the region')
plt.show()
```

```python
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('score range')
plt.ylabel('number of students')
plt.title('scores distribution of students')
plt.show()
```

```python
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies,exam_scores,color='green', marker='o', label='student data')

plt.xlabel('hours studies')
plt.ylabel('exam scores')
plt.title('relationship between study time and exam scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3], [50,55,60], color='blue',label='class a')
plt.scatter([1,2,3], [45,50,52], color='orange',label='class b')
plt.xlabel('hours studies')
plt.ylabel('exam scores')
plt.title('relationship between study time and exam scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line chart')
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize =(10,5))
x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y, color = 'red')
ax[0].set_title('lineplot')

ax[1].bar(x,y, color = 'green')
ax[1].set_title('bar chart')
plt.tight_layout()

plt.show()

```

```python
fig, ax = plt.subplots(1,2, figsize =(10,5))
x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y, color = 'red')
ax[0].set_title('lineplot')

ax[1].bar(x,y, color = 'green')
ax[1].set_title('bar chart')
plt.tight_layout()

fig.suptitle('comparison of line and bar charts')

plt.show()
```
